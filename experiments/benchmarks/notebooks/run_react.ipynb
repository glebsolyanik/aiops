```python
from aiopslab.orchestrator import Orchestrator  
from aiopslab.service.kubectl import KubeCtl 
from clients.react import Agent
```

```python
network_loss_problem_id = "network_loss_hotel_res-detection-1"
namespace_nl = "test-hotel-reservation"
focus_services_nl = ["frontend", "user", "search", "profile", "recommendation", "reservation"]

orch_nl = Orchestrator()
kubectl_nl = KubeCtl()
react_agent = Agent()

orch_nl.register_agent(react_agent, name="network-loss-research-agent")
```

```python
problem_desc_nl, instructs_nl, apis_nl = orch_nl.init_problem(network_loss_problem_id)
```

```python
react_agent.init_context(problem_desc_nl, instructs_nl, apis_nl)
```

```python
import asyncio
import wandb

wandb.init(project="AIOpsLab")

output = await orch_nl.start_problem(max_steps=30)
```

```python
# Все шаги диалога (agent и env по очереди)
for i, item in enumerate(output["history"]):
    role = item.role if hasattr(item, "role") else item.get("role")
    content = item.content if hasattr(item, "content") else item.get("content", "")
    print(f"--- Step {i+1} [{role}] ---")
    print(content[:500] + ("..." if len(content) > 500 else ""))
    print()

# Только ответы агента (assistant)
agent_messages = [h for h in output["history"] if (h.role if hasattr(h, "role") else h.get("role")) == "assistant"]
print(f"Всего ответов агента: {len(agent_messages)}")
for i, msg in enumerate(agent_messages):
    c = msg.content if hasattr(msg, "content") else msg.get("content", "")
    print(f"--- Ответ агента {i+1} ---\n{c[:800]}{'...' if len(c) > 800 else ''}\n")

# Сохранённое решение (то, что пошло в eval — от последнего submit)
print("Сохранённое решение (submit):", orch_nl.session.solution)
```
